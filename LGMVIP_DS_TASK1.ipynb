1.IMPORTING LIBRAIRES

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

2.LOADING DATASET

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
df = pd.read_csv('iris.data', names=columns)
df.head()
```

3.VISUALIZE THE DATASET

```python
# Some basic statistical analysis about the dataset
df.describe()
```

```python
# Visualize the whole dataset using seaborn pairplot
g=sns.pairplot(df, hue='Class_labels')
g.fig.set_figheight(6)
g.fig.set_figwidth(10)
```

4.SEPARATING INPUT AND OUTPUT COLUMN:

```python
data = df.values
X = data[:,0:4]
Y = data[:,4]
print(X)
print(Y)
```

5.SPLITING THE DATA INTO TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

MODEL 1: SUPPORT VECTOR MACHINE ALGORITHM

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
# Predict from the test dataset
predictions1 = model_svc.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions1)*100)
```

MODEL 2: LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
# Predict from the test dataset
predictions2 = model_lr.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions2)*100)
```

MODEL 3: DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc=LogisticRegression()
model_dtc.fit(X_train, y_train)
```

```python
# Predict from the test dataset
predictions3 = model_dtc.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions3)*100)
```

6. CLASSIFICATION REPORTS

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions1))
print('-' * 60)
print(classification_report(y_test,predictions2))
print('-' * 60)
print(classification_report(y_test,predictions3))
print('-' * 60)
```
